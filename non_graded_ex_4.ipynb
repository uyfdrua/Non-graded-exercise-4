```python
import random
import time

# Define the cards with their characteristics
card_deck = [
    {"Name": "Diablo", "Health": 100, "Attack": 90, "Defense": 60},
    {"Name": "Medusa", "Health": 100, "Attack": 70, "Defense": 70},
    {"Name": "Jester", "Health": 120, "Attack": 60, "Defense": 90},
    {"Name": "Troll", "Health": 150, "Attack": 40, "Defense": 94},
    {"Name": "Specter", "Health": 100, "Attack": 70, "Defense": 70},
    {"Name": "Mist", "Health": 100, "Attack": 75, "Defense": 65},
    {"Name": "Savage", "Health": 100, "Attack": 90, "Defense": 50},
    {"Name": "Marauder", "Health": 100, "Attack": 85, "Defense": 50},
    {"Name": "Wimp", "Health": 110, "Attack": 40, "Defense": 85},
    {"Name": "Sorcerer", "Health": 100, "Attack": 70, "Defense": 55}
]

# Function to initialize a new game
def initialize_game(deck_size):
    player_deck = random.sample(card_deck, deck_size)
    opponent_deck = random.sample(card_deck, deck_size)
    return player_deck, opponent_deck

# Function to print card details
def print_card(card):
    return f"Name: {card['Name']}\nHealth: {card['Health']}\nAttack: {card['Attack']}\nDefense: {card['Defense']}\n"

# Function to simulate a round of the game
def play_round(player_deck, opponent_deck):
    while player_deck and opponent_deck:
        # Display player cards
        print("Your cards:")
        for i, card in enumerate(player_deck, start=1):
            print(f"{i}. {card['Name']}")
        
        # Draw a random opponent card
        opponent_card = random.choice(opponent_deck)
        print(f"\nOpponent's card is:\n{print_card(opponent_card)}")
        
        # Ask player to choose a card
        choice = int(input("Pick a card from your deck (number): ")) - 1
        player_card = player_deck.pop(choice)
        print(f"\nYou chose:\n{print_card(player_card)}")
        
        # Determine the outcome of the fight
        if player_card['Attack'] > opponent_card['Health']:
            print(f"Your {player_card['Name']} defeats the opponent's {opponent_card['Name']}!\n")
            opponent_deck.remove(opponent_card)
        else:
            # Calculate damage
            defense_left = opponent_card['Defense'] - player_card['Attack']
            if defense_left < 0:
                damage = -defense_left
                opponent_card['Defense'] = 0
                opponent_card['Health'] -= damage
            else:
                opponent_card['Defense'] = defense_left
            
            # Display updated opponent card status
            print(f"\nOpponent's card status after your attack:\n{print_card(opponent_card)}")
            
            if opponent_card['Health'] <= 0:
                print(f"Your {player_card['Name']} defeats the opponent's {opponent_card['Name']}!\n")
                opponent_deck.remove(opponent_card)
            else:
                # Opponent's turn
                print(f"\nOpponent's turn. Opponent attacks your {player_card['Name']}!")
                opponent_attack = opponent_card['Attack']
                player_defense_left = player_card['Defense'] - opponent_attack
                if player_defense_left < 0:
                    damage = -player_defense_left
                    player_card['Defense'] = 0
                    player_card['Health'] -= damage
                else:
                    player_card['Defense'] = player_defense_left
                
                # Display updated player card status
                print(f"\nYour card status after the opponent's attack:\n{print_card(player_card)}")
                
                if player_card['Health'] <= 0:
                    print(f"Your {player_card['Name']} was defeated by the opponent's {opponent_card['Name']}!\n")
                    return False  # Player lost
                else:
                    print("Your card survived the attack!")

        time.sleep(3)  # Pause for 3 seconds before next round
    
    return len(opponent_deck) == 0  # Return True if opponent's deck is empty, False otherwise

# Initialize the game
deck_size = 5  # Number of cards in each deck
player_deck, opponent_deck = initialize_game(deck_size)

# Start the game
if play_round(player_deck, opponent_deck):
    print("Congratulations! You have defeated all opponent's cards!")
else:
    print("Game Over! The opponent defeated you.")

```
